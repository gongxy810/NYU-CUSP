CHART 1 Monthly deaths count time series for US combined, by race/ethnicity group

Data downloaded:

For mortality counts over time and for crude rates over time MONTHLY/NATIONAL - https://data.cdc.gov/NCHS/Provisional-COVID-19-Deaths-Distribution-of-Deaths/pj7m-y5uh

```python
data1 = pd.read_csv("8093_raw_data/Provisional_COVID-19_Deaths__Distribution_of_Deaths_by_Race_and_Hispanic_Origin_20240511.csv")
```

```python
data1.head()
```

```python
# Filter the data
mortality_data = data1[
    (data1['Group'] == 'By Month') &
    (data1['State'] == 'United States') &
    (data1['Indicator'] == 'Count of COVID-19 deaths')
]
```

```python
# Rename race/ethnicity columns for ease of reading
renaming_dict = {
    'Non-Hispanic White': 'White',
    'Non-Hispanic Black or African American': 'Black',
    'Non-Hispanic American Indian or Alaska Native': 'Indigenous',
    'Non-Hispanic Asian': 'Asian',
    'Non-Hispanic Native Hawaiian or Other Pacific Islander': 'Pacific Islander',
    'Non Hispanic more than one race': 'More than one race',
    'Hispanic or Latino': 'Latino'
}
mortality_data.rename(columns=renaming_dict, inplace=True)

mortality_data.head()
```

```python
# Create year-month column from the End Date
mortality_data['year_month'] = pd.to_datetime(mortality_data['End Date']).dt.to_period('M')

# Add a column for all races combined
race_columns = ['White', 'Black', 'Indigenous', 'Asian', 'Pacific Islander', 'More than one race', 'Latino']
mortality_data['US Total'] = mortality_data[race_columns].sum(axis=1)
```

```python
mortality_data.head()
```

```python
# Select just the columns needed for chart
final_columns = ['year_month'] + race_columns + ['US Total']
chart1 = mortality_data[final_columns]

# Display the final prepared data

chart1.head()
```

CHART 2
Monthly crude rates over time, adding in each month for cumulative rates, for US total

Data downloaded

For national populations by race, all ages combined, 2020 Census via CDC WONDER - https://wonder.cdc.gov/controller/saved/D170/D278F683

Also uses mortality data; same data as for Chart 1.

```python
# Specify the path to your file
file_path = "8093_raw_data/Single-Race Population Estimates 2010-2020 by State and Single-Year Age.txt"

# Load the text file, assuming it's tab-delimited
data2 = pd.read_csv(file_path, sep='\t')  # Use sep="," if it's comma-separated, or sep="\s+" for space-separated
```

```python
data2.head()
```

```python
# Filter for just population data that is needed: total Hispanic/Latino (all races) and non-Hispanic separated by race
pop_US = data2[((data2['Notes'] == "Total") & (data2['Ethnicity'] == "Hispanic or Latino")) | ((pd.isna(data2['Notes']) | (data2['Notes'] != "Total")) & (data2['Ethnicity'] != "Hispanic or Latino"))]

# Mutate Race.Ethnicity and select columns
pop_US['Race.Ethnicity'] = pop_US['Race'].combine_first(pop_US['Ethnicity'])
pop_US = pop_US[['Race.Ethnicity', 'Population']]
pop_US['State'] = "United States"

# Rename race/ethnicity
pop_US['Race.Ethnicity'] = pop_US['Race.Ethnicity'].replace({"Black or African American": "Black", 
                                                             "American Indian or Alaska Native": "Indigenous",
                                                             "Native Hawaiian or Other Pacific Islander": "Pacific Islander",
                                                             "Hispanic or Latino": "Latino"})
```

```python
# Health data - same as used for chart 1


# Filter deaths over time to include just national data - same data as chart 1
chart_4 = mortality_data[(mortality_data['Group'] == "By Month") & (mortality_data['Indicator'] == "Count of COVID-19 deaths") & (mortality_data['State'] == "United States")]

# Rename race/ethnicity variables
chart_4 = chart_4.rename(columns={"Non-Hispanic White": "White", 
                                  "Non-Hispanic Black or African American": "Black", 
                                  "Non-Hispanic Asian": "Asian", 
                                  "Non-Hispanic American Indian or Alaska Native": "Indigenous", 
                                  "Non-Hispanic Native Hawaiian or Other Pacific Islander": "Pacific Islander", 
                                  "Non-Hispanic more than one race": "More than one race", 
                                  "Hispanic or Latino": "Latino"})

# Make data long instead of wide and get rid of unneeded columns
chart_4 = chart_4.melt(id_vars=['End Date', 'Year', 'Month', 'State'], 
                       value_vars=['White', 'Black', 'Indigenous', 'Asian', 'Pacific Islander', 'Latino', 'More than one race'], 
                       var_name='Race.Ethnicity', 
                       value_name='COVID.19.Deaths')
chart_4 = chart_4[['End Date', 'Year', 'Month', 'State', 'Race.Ethnicity', 'COVID.19.Deaths']]
```

```python
# Format 'End Date' as a date
chart_4['End Date'] = pd.to_datetime(chart_4['End Date'])
```

```python
chart_4.head()
```

```python
# Merge chart_4 and pop_US, specifying suffixes for overlapping columns
chart_4 = pd.merge(chart_4, pop_US, on='Race.Ethnicity', how='left', suffixes=('_chart', '_pop'))

# Display the first few rows to verify the changes
chart_4.head()
```

```python
chart_4['Cumulative_Total'] = chart_4.groupby('Race.Ethnicity')['COVID.19.Deaths'].cumsum()
```

```python
# Calculate cumulative crude rates by month
chart_4['Cumulative_Crude'] = chart_4['Cumulative_Total'] / chart_4['Population'] * 100000
```

```python
chart_4.head()
```

```python
# Check the data types of each column
print(chart_4.dtypes)
```

```python
# Convert Year and Month columns to strings and concatenate them with appropriate formatting
chart_4['YearMonth'] = chart_4['Month'].astype(int).map("{:02d}".format) + '/' + chart_4['Year'].astype(int).astype(str)
chart_4.head()
```

```python
# Select/rearrange data
chart_4 = chart_4[['End Date', 'State_chart', 'Race.Ethnicity', 'COVID.19.Deaths', 'Cumulative_Total', 'Cumulative_Crude', 'YearMonth']]

chart_4.head()
```

```python
# Select/rearrange data
chart_4 = chart_4[['YearMonth', 'State_chart', 'Race.Ethnicity', 'Cumulative_Crude']]

# Pivot the table
chart_2 = chart_4.pivot_table(index=['YearMonth', 'State_chart'], columns='Race.Ethnicity', values='Cumulative_Crude')

chart_2.head()
```

```python
chart_2.to_csv("chart_2.csv")
```

```python
import pandas as pd

# Filter mortality data by age groups for States
mort_by_ages_US = df4[df4['AgeGroup'].isin(["All ages, standardized", "All ages, unadjusted"]) & (df4['Race/Hispanic origin'] != "Other")]
mort_by_ages_US = mort_by_ages_US[['Data as of', 'End Date', 'State', 'Race/Hispanic origin', 'Count of COVID-19 deaths', 'AgeGroup']]
mort_by_ages_US = mort_by_ages_US.rename(columns={"Race/Hispanic origin": "Race.Ethnicity"})
mort_by_ages_US = mort_by_ages_US.rename(columns={"Count of COVID-19 deaths": "Count.of.COVID.19.deaths"})


# Replace race/ethnicity names
race_ethnicity_mapping = {
    "Non-Hispanic White": "White",
    "Non-Hispanic Black": "Black",
    "Non-Hispanic Asian": "Asian",
    "Non-Hispanic American Indian or Alaska Native": "Indigenous",
    "Non-Hispanic Native Hawaiian or Other Pacific Islander": "Pacific Islander",
    "Hispanic": "Latino"
}
mort_by_ages_US['Race.Ethnicity'] = mort_by_ages_US['Race.Ethnicity'].replace(race_ethnicity_mapping)

# Replace age group names
age_group_mapping = {"85 years and over": "85+ years"}
mort_by_ages_US['AgeGroup'] = mort_by_ages_US['AgeGroup'].replace(age_group_mapping)


mort_by_ages_US.head()
```

```python
# Combine the NYC and NYS values
mort_by_ages_NY = mort_by_ages_US[mort_by_ages_US['State'].isin(["New York", "New York City"])]
mort_by_ages_NY_new = mort_by_ages_NY.groupby(['Race.Ethnicity', 'AgeGroup'])['Count.of.COVID.19.deaths'].sum().reset_index()

# Rename the 'Count.of.COVID.19.deaths' column to 'Count.of.COVID.19.deaths_COMBO'
mort_by_ages_NY_new_renamed = mort_by_ages_NY_new.rename(columns={"Count.of.COVID.19.deaths": "Count.of.COVID.19.deaths_COMBO"})
```

```python
# Remove NY data from the original dataset
mort_by_ages_NOT_NY = mort_by_ages_US[~mort_by_ages_US['State'].isin(["New York", "New York City"])]

# Merge the non-NY dataset with the new NY data
mort_by_ages_US = pd.concat([mort_by_ages_NOT_NY, mort_by_ages_NY_new_renamed])

# See which states/groups could be directly age adjusted
mort_by_ages_US_NAcount = mort_by_ages_US.groupby(['State', 'Race.Ethnicity']).apply(lambda x: sum(x['Count.of.COVID.19.deaths'].isna()))
mort_by_ages_US_NAcount = mort_by_ages_US_NAcount.reset_index(name='sumNA').query('sumNA == 0')


mort_by_ages_US_NAcount.head()
```

```python
# Use merge to select data with 0 NAs
mort_by_ages_US_no_NAs = pd.merge(mort_by_ages_US_NAcount, mort_by_ages_US, on=['State', 'Race.Ethnicity'], how='left')

# Merge with population data
mort_by_ages_US_no_NAs = pd.merge(mort_by_ages_US_no_NAs, age_groups_all_US, on=['State', 'Race.Ethnicity'], how='left')

# Calculate crude rate per age/race/state group
mort_by_ages_US_no_NAs['crude_rate_group'] = mort_by_ages_US_no_NAs['Count.of.COVID.19.deaths'] / mort_by_ages_US_no_NAs['Population'] * 100000

mort_by_ages_US_no_NAs.head()
```

```python
# Define a function to choose between direct and indirect rates
def choose_rate(row):
    if not np.isnan(row['age_adjusted_rate_direct']):
        return row['age_adjusted_rate_direct']
    else:
        return row['age_adjusted_rate_indirect']

# Apply the function to create the new column
chart_6_final['Age_Adjusted_Rate_GRAPHIC'] = chart_6_final.apply(choose_rate, axis=1)
```

```python
# Convert 'Count.of.COVID.19.deaths' column to numeric
chart_6_final['Count.of.COVID.19.deaths'] = pd.to_numeric(chart_6_final['Count.of.COVID.19.deaths'], errors='coerce')

# Add commas to COVID death counts
chart_6_final['Count.of.COVID.19.deaths'] = chart_6_final['Count.of.COVID.19.deaths'].apply(lambda x: "{:,.0f}".format(x) if pd.notnull(x) else "")

# Round Crude_Cumulative and Age_Adjusted_Rate_GRAPHIC
chart_6_final['Crude_Cumulative'] = chart_6_final['Crude_Cumulative'].round()
chart_6_final['Age_Adjusted_Rate_GRAPHIC'] = chart_6_final['Age_Adjusted_Rate_GRAPHIC'].round()
```

```python
# Rename variables and select necessary columns
chart_6_final = chart_6_final.rename(columns={"Crude_Cumulative": "Crude Rate", 
                                              "Age_Adjusted_Rate_GRAPHIC": "Age-Adjusted Rate", 
                                              "Count.of.COVID.19.deaths": "Number of COVID-19 Deaths"})

chart_6_final = chart_6_final[["End.Date", "State", "Race.Ethnicity", "Number of COVID-19 Deaths", 
                                "Population", "Crude Rate", "age_adjusted_rate_direct", 
                                "age_adjusted_rate_indirect", "Direct_Available", "Age-Adjusted Rate"]]


chart_6_final.head()
```
